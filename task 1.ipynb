```python
#task 1
raw_products = [
    ("Laptop", "1200", "Electronics"),
    ("Phone", "800", "Electronics"),
    ("Laptop", "1200", "Electronics"),
    ("Table", "150", "Furniture"),
    ("Chair", None, "Furniture"),
    ("Pen", "2", "Stationery")
]
clean_products = []
seen = set()

for name, price, category in raw_products:

    if price is None:
        continue

    price = float(price)

    product_key = (name, price, category)

    if product_key in seen:
        continue
    seen.add(product_key)

    clean_products.append({
        "name": name,
        "price": price,
        "category": category
    })

catalogue = {}

for product in clean_products:

    category = product["category"]

    if category not in catalogue:
        catalogue[category] = []

    catalogue[category].append(product)

print("\nProduct Summary\n")

for category, items in catalogue.items():

    print(f"Category: {category}")

    for item in items:
        print(f"  {item['name']} - ${item['price']}")

    print()
```
