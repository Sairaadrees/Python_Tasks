```python
class BankAccount:

    bank_name = "Secure Bank"

    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):

        if amount <= 0:
            print("Invalid deposit amount")
            return

        self.__balance += amount
        print("Deposit successful")

    def withdraw(self, amount):

        if amount <= 0:
            print("Invalid withdrawal amount")
            return

        if amount > self.__balance:
            print("Insufficient funds")
            return

        self.__balance -= amount
        print("Withdrawal successful")

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, value):
        print("Balance cannot be set directly")

    @classmethod
    def bank_info(cls):
        print("Bank Name:", cls.bank_name)

    @staticmethod
    def validate_amount(amount):
        return amount > 0

    def __str__(self):
        return f"Account Owner: {self.owner}, Balance: {self.__balance}"

    def __repr__(self):
        return f"BankAccount('{self.owner}', {self.__balance})"


account1 = BankAccount("Alice", 1000)

print(account1)

account1.deposit(500)
account1.withdraw(200)

print("Current balance:", account1.balance)

BankAccount.bank_info()

print("Amount valid:", BankAccount.validate_amount(100)
     )
```
