```python
import asyncio
import os
from dotenv import load_dotenv
from datetime import datetime

# Load secrets from .env
load_dotenv()
SOURCE1_API_KEY = os.getenv("SOURCE1_API_KEY", "key1")
SOURCE2_API_KEY = os.getenv("SOURCE2_API_KEY", "key2")
SOURCE3_API_KEY = os.getenv("SOURCE3_API_KEY", "key3")

# Simulate a data source fetch (async)
async def fetch_source1():
    await asyncio.sleep(1)  # simulate network delay
    return [
        {"id": 1, "name": "Alice", "score": 85},
        {"id": 2, "name": "Bob", "score": 90},
    ]

async def fetch_source2():
    await asyncio.sleep(2)
    return [
        {"id": 3, "name": "Charlie", "score": 75},
        {"id": 4, "name": "David", "score": 95},
    ]

async def fetch_source3():
    await asyncio.sleep(1.5)
    return [
        {"id": 5, "name": "Emma", "score": 80},
        {"id": 2, "name": "Bob", "score": 92},  # duplicate ID to test merging
    ]

# Merge and summarize data
def merge_data(*sources):
    merged = {}
    for source in sources:
        for record in source:
            # if duplicate, keep highest score
            if record["id"] not in merged or record["score"] > merged[record["id"]]["score"]:
                merged[record["id"]] = record
    return list(merged.values())

# Print summary
def print_summary(data):
    print("\n--- Combined Data Summary ---")
    for record in sorted(data, key=lambda x: x["score"], reverse=True):
        print(f"{record['name']} (ID: {record['id']}) -> Score: {record['score']}")

# Main async function
async def main():
    results = await asyncio.gather(
        fetch_source1(),
        fetch_source2(),
        fetch_source3()
    )
    combined = merge_data(*results)
    print_summary(combined)

# Run in Jupyter-friendly way
await main()
```
