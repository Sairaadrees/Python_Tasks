```python
import csv
csv_files = [
    "students_clean.csv",
    "students_invalid.csv",
    "students_corrupted.csv"
]

valid_students = []

for filename in csv_files:
    try:
        with open(filename, "r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                try:
                    required_cols = ["student_id","name","course","marks","attendance","email"]
                    if not all(col in row for col in required_cols):
                        raise ValueError("Missing columns")

                    marks = int(row["marks"])
                    attendance = int(row["attendance"])

                    if marks < 0 or marks > 100 or attendance < 0 or attendance > 100:
                        raise ValueError("Impossible marks or attendance")
                    valid_students.append({
                        "student_id": row["student_id"],
                        "name": row["name"],
                        "course": row["course"],
                        "marks": marks,
                        "attendance": attendance,
                        "email": row["email"]
                    })
                except ValueError as ve:
                    print(f"Skipping invalid row in {filename}: {row} -> {ve}")
    except FileNotFoundError:
        print(f"File not found: {filename}")
    except Exception as e:
        print(f"Error reading file {filename}: {e}")

output_file = "cleaned_student_results.csv"
with open(output_file, "w", newline="") as f:
    fieldnames = ["student_id","name","course","marks","attendance","email"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for student in valid_students:
        writer.writerow(student)

print("\nFailed Students (marks < 50):")
for student in valid_students:
    if student["marks"] < 50:
        print(f"{student['name']} - Marks: {student['marks']}")
```
