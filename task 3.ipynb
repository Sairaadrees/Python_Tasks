```python
# task 3
from auditor_module import scan_folder, files_older_than, calculate_penalties
from datetime import datetime
import csv

project_folder = r"C:\Users\PMLS\Documents"  
```

```python
# Scan  files
files_info = scan_folder(project_folder)

old_files = files_older_than(files_info, days=30)

penalties = calculate_penalties(old_files, fine_per_file=10)
```

```python

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
audit_file = f"audit_report_{timestamp}.csv"
penalty_file = f"penalty_report_{timestamp}.csv"

with open(audit_file, "w", newline="") as f:
    fieldnames = ["path","name","size_bytes","type","last_modified"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for fi in files_info:
        writer.writerow({
            "path": fi["path"],
            "name": fi["name"],
            "size_bytes": fi["size_bytes"],
            "type": fi["type"],
            "last_modified": fi["last_modified"].strftime("%Y-%m-%d %H:%M:%S")
        })

with open(penalty_file, "w", newline="") as f:
    fieldnames = ["file","penalty"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for p in penalties:
        writer.writerow(p)
```

```python
print(f"Audit report saved: {audit_file}")
print(f"Penalty report saved: {penalty_file}")
print(f"Files older than 30 days: {len(old_files)}")
```
