```python
#task(part 1)
#task 4
from collections import defaultdict
products = [
    {"name": "Laptop", "price": "1200", "category": "Electronics"},
    {"name": "Phone", "price": None, "category": "Electronics"},
    {"name": "Desk", "price": "200", "category": "Furniture"},
    {"name": "Chair", "price": "150", "category": "Furniture"},
    {"name": "Laptop", "price": "1200", "category": "Electronics"},  # duplicate
    {"name": "Headphones", "price": "100", "category": "Electronics"},
    {"name": "Table", "price": "300", "category": "Furniture"},
    {"name": "Monitor", "price": "450", "category": "Electronics"},
]
seen = set()
cleaned_products = [
    {
        "name": p["name"],
        "price": float(p["price"]),
        "category": p["category"]
    }
    for p in products
    if p["price"] not in (None, "")
    and (p["name"], p["price"], p["category"]) not in seen
    and not seen.add((p["name"], p["price"], p["category"]))
]

categories = defaultdict(list)

for p in cleaned_products:
    categories[p["category"]].append(p)

summary = {
    category: sorted(items, key=lambda x: x["price"], reverse=True)[:5]
    for category, items in categories.items()
}
for category, items in summary.items():
    print(f"\nCategory: {category}")
    for item in items:
        print(f"{item['name']} - ${item['price']}")
```

```python
#part 2
employees = [
    {"name": "Alice", "performance": 95, "experience": 5},
    {"name": "Bob", "performance": 95, "experience": 7},
    {"name": "Charlie", "performance": 88, "experience": 3},
    {"name": "Dave", "performance": 95, "experience": 7},
]

ranked = sorted(
    employees,
    key=lambda x: (-x["performance"], -x["experience"], x["name"])
)

employee_lookup = {e["name"]: e for e in ranked}

print("Search result for Bob:", employee_lookup.get("Bob"))
print("\nRanked Employees:")
for i, e in enumerate(ranked, 1):
    print(f"{i}. {e['name']} - Perf: {e['performance']}, Exp: {e['experience']}")
```
