```python
#task 5
from functools import reduce

orders = [
    {"customer": "Alice", "category": "Electronics", "amount": 1200},
    {"customer": "Bob", "category": "Clothing", "amount": 200},
    {"customer": "Charlie", "category": "Electronics", "amount": 800},
    {"customer": "Dave", "category": "Groceries", "amount": 90},
    {"customer": "Eva", "category": "Clothing", "amount": 400},
]

discount_rules = {
    "Electronics": 0.10,
    "Clothing": 0.20,
    "Groceries": 0.05
}

MIN_ORDER = 100

apply_discount = lambda o: {
    "customer": o["customer"],
    "category": o["category"],
    "final_amount": o["amount"] * (1 - discount_rules.get(o["category"], 0))
}
discounted_orders = list(map(apply_discount, orders))

#Filter orders below minimum threshold
valid_orders = list(filter(lambda o: o["final_amount"] >= MIN_ORDER, discounted_orders))
# Sort orders by highest value
sorted_orders = sorted(valid_orders, key=lambda x: x["final_amount"], reverse=True)

total_revenue = reduce(lambda acc, o: acc + o["final_amount"], sorted_orders, 0)
summary = {
    "processed_orders": sorted_orders,
    "total_revenue": total_revenue,
    "order_count": len(sorted_orders)
}

print(summary)
```
