```python
from collections import defaultdict
from itertools import islice

def stream_errors(file_path):
    with open(file_path, "r") as f:
        for line in f:  
            if "ERROR" in line or "CRITICAL" in line:
                yield line.strip()

def extract_error_type(log_line):
    parts = log_line.split()
    for p in parts:
        if "Error" in p or "ERROR" in p or "CRITICAL" in p:
            return p
    return "Unknown"

def process_log_stream(file_path):

    error_counts = defaultdict(int)

    for log in stream_errors(file_path):

        error_type = extract_error_type(log)

        error_counts[error_type] += 1

        print(f"{error_type} → count: {error_counts[error_type]}")

process_log_stream(r"C:\Users\PMLS\Documents\mydocuments\server.log.txt")
```
