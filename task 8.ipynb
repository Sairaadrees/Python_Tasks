```python
from abc import ABC, abstractmethod
from datetime import datetime


class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.borrowed = False

    def __str__(self):
        return self.title + " by " + self.author

    def __eq__(self, other):
        return self.title == other.title and self.author == other.author


class Member(ABC):

    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.borrowed:
            print("Book already borrowed")
            return

        if len(self.borrowed_books) >= self.limit():
            print("Borrow limit reached")
            return

        book.borrowed = True
        self.borrowed_books.append((book, datetime.now()))
        print(self.name, "borrowed", book)

    def return_book(self, book):
        for b, date in self.borrowed_books:
            if b == book:
                self.borrowed_books.remove((b, date))
                book.borrowed = False

                days = (datetime.now() - date).days
                fine = self.calculate_fine(days)

                print(self.name, "returned", book)
                return fine

        print("Book not found")
        return 0

    def __len__(self):
        return len(self.borrowed_books)

    def __getitem__(self, index):
        return self.borrowed_books[index][0]

    @abstractmethod
    def limit(self):
        pass

    @abstractmethod
    def calculate_fine(self, days):
        pass


class BasicMember(Member):

    def limit(self):
        return 2

    def calculate_fine(self, days):
        if days > 7:
            return (days - 7) * 2
        return 0


class PremiumMember(BasicMember):

    def limit(self):
        return 5

    def calculate_fine(self, days):
        if days > 14:
            return (days - 14) * 1
        return 0


class EliteMember(PremiumMember):

    def limit(self):
        return 10

    def calculate_fine(self, days):
        return 0


book1 = Book("Python Basics", "John Smith")
book2 = Book("AI Fundamentals", "Jane Doe")

member = PremiumMember("Alice")

member.borrow_book(book1)
member.borrow_book(book2)

print("Borrowed books:", len(member))
print("First book:", member[0])

fine = member.return_book(book1)
print("Fine:", fine)
```
